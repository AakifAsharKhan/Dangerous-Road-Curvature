```python
import numpy as np
import pandas as pd
```

```python
DATA_PATH = "E:\\MAHE\\1st Sem\\Projects\\Dangerous Road Curvature\\Data\\"
DATASET = DATA_PATH + "1st Data Collection.xlsx"
```

```python
df = pd.read_excel(DATASET)
df.head()
```

```python
df.shape
```

```python
df_1 = df[['time', 'Accr_x', 'Accr_y', 'Accr_z', 'Rot_x', 'Rot_z']]
print(df_1.shape)
df_1.head()
```

```python
df_2 = df[['Latitude', 'Longitude', 'gps_time']]
print(df_2.shape)
df_2.head()
```

```python
df_2 = df_2.dropna()
```

```python
df_2
```

```python
df_2.drop(labels = [0,1,2,3,4,5], axis=0, inplace = True)
df_2.shape
```

```python
li = []
for i in range(len(df_2)):
    if i == len(df_2) - 1:
        break
    li.append(df_2.iloc[i+1]['gps_time'] - df_2.iloc[i]['gps_time'])
print(len(li))
```

```python
df_2.drop([len(df_2)-1], axis = 0, inplace = True)
```

```python
df_2['time_difference'] = li
```

```python
df_2
```

```python
df_2['gps_time'] = df_2['gps_time'].round(3)
df_2['time_difference'] = df_2['time_difference'].round(3)
df_2
```

```python
df_2.reset_index(inplace=True)
```

```python
df_2
```

```python
df_2.drop('index', axis=1, inplace=True)
```

```python
df_2
```

```python
df_1['time'] = df_1['time'].round(3)
```

```python
df_2.loc[0]
```

```python
print(df_2,"\n", df_1)
```

```python
df_1['Magnitude_Accelerometer'] = np.sqrt(df_1['Accr_x']**2 + df_1['Accr_y']**2 + df_1['Accr_z']**2)
df_1['Magnitude_Gyroscope'] = np.sqrt(df_1['Rot_x']**2 + df_1['Rot_z']**2)
```

```python
temp1 = df_2.loc[0,'gps_time'].round(2)
temp2 = df_2.loc[1, 'gps_time'].round(2)
print(temp1, temp2)
```

```python
df_1[(df_1['time'] > temp1) | (df_1['time'] < temp2)]
```

```python
df_1[(df_1['time'] > temp1) & (df_1['time'] < temp2)]
```

```python
df_li = []
neww_df = pd.DataFrame(columns=['time', 'Accr_x', 'Accr_y', 'Accr_z', 'Magnitude_Accelerometer','Magnitude_Gyroscope','Latitude','Longitude'])
for i in range(len(df_2) - 1):
    temp_accel_dict = df_1[(df_1['time'] > df_2.loc[i,'gps_time']) & (df_1['time'] < df_2.loc[i+1,'gps_time'])].to_dict()
    temp_gps_dict = {'Latitude':np.repeat(df_2.loc[i,'Latitude'],len(temp_accel_dict['time'])), 
                     'Longitude':np.repeat(df_2.loc[i,'Longitude'],len(temp_accel_dict['time']))}

    temp_gps_df = pd.DataFrame(temp_gps_dict)
    temp_gps_df.reset_index(inplace=True)
    temp_gps_df.drop(['index'], axis=1, inplace=True)
    # print(temp_gps_df, "\n\n")
    temp_accel_df = pd.DataFrame(temp_accel_dict)
    temp_accel_df.reset_index(inplace=True)
    temp_accel_df.drop(['index'], axis=1, inplace=True)
    # print(temp_accel_df)
    df_li.append(pd.concat([temp_accel_df, temp_gps_df], axis=1))

output_df = pd.concat(df_li)
output_df.reset_index(inplace=True)
output_df.drop(['index'], axis=1, inplace=True)
output_df.to_csv('output.csv')
```
